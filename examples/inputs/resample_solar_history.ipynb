# Create a solar input file

Create a solar input file using available data using example data from the flatirons campus.

```python
import pandas as pd
```

```python
df_solar = pd.read_csv("Flatirons_solar_data_sunset_1s.csv")

# Set up the time_utc column, then drop Timestamp
df_solar["time_utc"] = pd.to_datetime(df_solar["Timestamp"], format="ISO8601", utc=True)
df_solar = df_solar.drop(columns=["Timestamp"])

# Add a time column where 0 is time 0
# s(note this will not be time synced with wind, just for example)
df_solar["time"] = (df_solar["time_utc"] - df_solar["time_utc"].min()).dt.total_seconds()



df_solar = df_solar.reset_index(drop=True)


df_solar
```

```python
# Save the data to a new CSV file
df_solar.to_pickle("solar_input.p")
```
